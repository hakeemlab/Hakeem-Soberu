```python
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
path= '/content/drive/MyDrive/Github Project/StudentPerformance.csv'
df=pd.read_csv(path)
df.head(11)
```

```python
# histogram showing distribution of math score
plt.hist(df["math score"], bins=4)
plt.title("Distribution of Math Score", fontsize= 14)
plt.show()
```

```python
#bar chart showing Average Reading Score by Gender
mean_reading_score_by_gender = df.groupby ("gender")["reading score"].mean()
mean_reading_score_by_gender
print(type(mean_reading_score_by_gender))

mean_reading_score_by_gender = pd.DataFrame(mean_reading_score_by_gender).reset_index()
print(type(mean_reading_score_by_gender))
print(mean_reading_score_by_gender)

plt.bar(mean_reading_score_by_gender["gender"],mean_reading_score_by_gender["reading score"])
plt.title("Average Reading Score by Gender")
plt.xlabel("reading score", fontsize=12)
plt.ylabel("gender", fontsize=12)
plt.show()
```

```python
# bar chart Average Math Score by Gender
mean_math_score_by_gender = df.groupby ("gender")["math score"].mean()
mean_math_score_by_gender
print(type(mean_math_score_by_gender))

mean_math_score_by_gender = pd.DataFrame(mean_math_score_by_gender).reset_index()
print(type(mean_math_score_by_gender))
print(mean_math_score_by_gender)

plt.bar(mean_math_score_by_gender["gender"],mean_math_score_by_gender["math score"])
plt.title("Average Math Score by Gender")
plt.xlabel("math score", fontsize=12)
plt.ylabel("gender", fontsize=12)
plt.show()
```

```python
from operator import index
# A side-by-side clustered bar chart showing gender by race/ethnicity by writing score
pd.pivot_table(df, index= "gender", columns= "race/ethnicity", values= "writing score", aggfunc= "mean")
pd.pivot_table(df, index="gender", columns="race/ethnicity", values="writing score", aggfunc= "mean").plot(kind= "bar", figsize=(6,4))
plt.xlabel("gender")
plt.ylabel("writing score")
plt.xticks(rotation=0)
plt.show()
```

```python
from operator import index
# A side-by-side clustered bar chart showing gender by race/ethnicity by reading
pd.pivot_table(df, index= "gender", columns= "race/ethnicity", values= "reading score", aggfunc= "mean")
pd.pivot_table(df, index="gender", columns="race/ethnicity", values="reading score", aggfunc= "mean").plot(kind= "bar", figsize=(6,4))
plt.xlabel("gender")
plt.ylabel("reading score")
plt.xticks(rotation=0)
plt.show()
```

```python
# A scatter plot showing math score and parental level of education
plt.scatter(df["math score"],df["parental level of education"])
plt.title("math score vs. parental level of education")
plt.xlabel("math score")
plt.ylabel("parental level of education")
plt.show()
```

```python
# A scatter plot showing math score and parental level of education
plt.scatter(df["writing score"],df["parental level of education"])
plt.title("writing score vs. parental level of education")
plt.xlabel("writing score")
plt.ylabel("parental level of education")
plt.show()
```

```python
# A scatter plot showing math score and parental level of education
plt.scatter(df["race/ethnicity"],df["reading score"], c= "green")
plt.title("race/ethnicity vs. reading score")
plt.xlabel("race/ethnicity")
plt.ylabel("reading score")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
path='/content/drive/MyDrive/Github Project/Salary_Data.csv'
df=pd.read_csv(path)
df.head(10)
```

```python
# A scatter plot showing years of exprience by salary greater than 40000
df2 = df[df.Salary>40000]
plt.scatter(df2 ["YearsExperience"], df2 ["Salary"], c = "green")
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience v Salary')
```

```python
mean_Salary_by_Years_experience= df.groupby("Salary") ["YearsExperience"].mean()
mean_Salary_by_Years_experience
print(type(mean_Salary_by_Years_experience))

mean_Salary_by_Years_experience = pd.DataFrame(mean_Salary_by_Years_experience).reset_index()
print(type(mean_Salary_by_Years_experience))
print(mean_Salary_by_Years_experience)

plt.bar(mean_Salary_by_Years_experience["YearsExperience"], mean_Salary_by_Years_experience["Salary"])
plt.title("Average Salary by Years of Experience", fontsize= 14)
plt.xlabel("YearsExperience", fontsize= 12)
plt.ylabel("Salary", fontsize= 12)
plt.show()
```

```python
#Trend of Years of Experience by Salary
plt.plot(df['YearsExperience'], df['Salary'], color= "red", marker = "o")
plt.title('Trend of YearsExperience by Salary', fontsize= 14)
plt.xlabel('YearsExperience', fontsize= 12)
plt.ylabel('Salary', fontsize= 12)
```
