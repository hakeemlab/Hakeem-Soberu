```python
from typing import DefaultDict
path="/content/drive/MyDrive/MSDA 673 Data/Life Expectancy.csv"
df= pd.read_csv(path)
df
```

```python
# Creat a histogram showing the distribution of Life Expectancy
plt.hist(df["Life Expectancy"], bins = 4)
# change the number of bins into 4
# Add title; fontsize = 14
plt.title("Distribution of Life of Expectancy", fontsize = 14)
# Add labels on x-axis and y-axis; fontsize = 12
# show the plot
plt.show()
```

```python
# Create a column bar chart showing average Life Expectancy by Status
mean_Life_Expectancy_by_Status = df.groupby("Status")["Life Expectancy"].mean()
mean_Life_Expectancy_by_Status
print(type(mean_Life_Expectancy_by_Status))

mean_Life_Expectancy_by_Status = pd.DataFrame(mean_Life_Expectancy_by_Status).reset_index()
print(type(mean_Life_Expectancy_by_Status))
print(mean_Life_Expectancy_by_Status)

plt.bar(mean_Life_Expectancy_by_Status["Status"],mean_Life_Expectancy_by_Status["Life Expectancy"])
# Add title; fontsize = 14
plt.title("Average Life Expectancy by Status", fontsize = 12)
# Add labels on x-axis and y-axis; fontsize = 12
plt.xlabel("Life Expectancy", fontsize = 12)
plt.ylabel("Status", fontsize = 12)
# show the plot
plt.show()
```

```python
# Create a horizontal bar chart showing the numbers of countires by Status (developed and developing)
Status_count = df["Status"].value_counts()
Status_count

plt.barh(Status_count.index,Status_count.values)
# Add title; fontsize = 14
plt.title("Number of Countries by Status", fontsize = 14)
# Add labels on x-axis and y-axis; fontsize = 12
plt.xlabel("No of Countries")
plt.ylabel("Status")
# show the plot
plt.show()
```

```python
# Create a scatter plot showing the relationship between Schooling and Life Expectancy
plt.scatter(df["Schooling"], df["Life Expectancy"], c = "green")
# Change the dot color into green
# Add title; fontsize = 14
plt.title("Schooling vs. Life Expectancy")
# Add labels on x-axis and y-axis; fontsize = 12
plt.xlabel("Schooling")
plt.ylabel("Life Expectancy")
# show the plot
plt.show()
```

```python
# A side-by-side (clustered) bar chart showing Average Life Expectancy by Status by Continent
from operator import index
print(df)
pd.pivot_table(df, index= "Continent", columns= "Status", values= "Life Expectancy", aggfunc= "mean")
pd.pivot_table(df, index= "Continent", columns= "Status", values= "Life Expectancy", aggfunc= "mean").plot(kind= "bar")
# Add title; fontsize = 14
plt.title("Life Expectancy by Status by Continent")
# Add labels on x-axis and y-axis; fontsize = 12
plt.xlabel("Life Expectancy", fontsize = 12)
plt.ylabel("Status", fontsize = 12)
# place the legend on the right side of the plot; adjust the place by changing the value in bbox_to_anchor()
plt.legend(["Status"],loc= "center right", bbox_to_anchor=(1.3,0.05))
# show the plot
plt.show()
```

```python
# Create a stacked bar chart showing number of countries by Status by Continent
pd.pivot_table(df, index= "Status", columns= "Continent", values= "Country", aggfunc= "count").plot(kind = "bar", stacked = True)
# Add title; fontsize = 14
plt.title("Number of Countries by Status by Continent", fontsize = 14)
# Add labels on x-axis and y-axis; fontsize = 12
plt.xlabel("Status")
plt.ylabel("Continent")
# show the plot
plt.show()
```

```python
# Create a pie chart showing number of countries by Continent
Continent= "1", "2", "3", "4", "5", "6", "7", "8"
# hint: replace sum() with count() in the groupby()
Countries_by_Continent = df.groupby(["Continent"])["Country"].count()
plt.pie(Countries_by_Continent, labels=Continent)
# Add title; fontsize = 14
plt.title("Number of Countries by Continent")
# show the plot
plt.show()
```
