```python
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
path= '/content/drive/MyDrive/Github Project/50_Startups.csv'
df=pd.read_csv(path)
df.head(11)
```

```python
# A histogram showing total counts of profits
plt.hist(df["Profit"], bins =11)
plt.title("Profits by counts", fontsize = 14)
plt.xlabel("Profit", fontsize = 12)
plt.ylabel("count", fontsize = 12)
```

```python
import seaborn as sns
```

```python
#Trend showing Marketing Spend by Density
df= df.dropna(subset=['Marketing Spend'])
sns.distplot(df["Marketing Spend"], bins =7)
```

```python
#Trend of Years of Experience by Salary
plt.plot(df['R&D Spend'], df['Profit'], color= "green", marker = "o")
plt.title('Trend of R&D Spend by profit', fontsize= 14)
plt.xlabel('R&D Spend', fontsize= 12)
plt.ylabel('Profit', fontsize= 12)
```

```python
# A scatter plot showing R&D Spend by Marketing greater than or equal to 100000
df[df.Administration>=100000]
plt.scatter(df["R&D Spend"], df ["Marketing Spend"], c = "green")
plt.xlabel('R&D Spend')
plt.ylabel('Marketing Spend')
plt.title('R&D Spend vs Marketing Spend')
```

```python
# A scatter plot showing Marketing Spend by Marketing greater than or equal to 1500000
df[df.Administration<=1500000]
plt.scatter(df["Marketing Spend"], df ["Profit"], c = "green")
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.title('Marketing Spend vs Profit')
```

```python
# pie chart showing percentage of States by Profit
State= "California", "Florida", "New York"
State_by_Profit = df.groupby(["State"]) ["Profit"].count()
plt.pie(State_by_Profit, labels=State, autopct= '%2.0f%%')
plt.title("States by Profits")
plt.show
```

```python
# bar chart Mean Profit by State
mean_Profit_by_State = df.groupby ("State")["Profit"].mean()
mean_Profit_by_State
print(type(mean_Profit_by_State))

mean_Profit_by_State = pd.DataFrame(mean_Profit_by_State).reset_index()
print(type(mean_Profit_by_State))
print(mean_Profit_by_State)

plt.bar(mean_Profit_by_State["State"],mean_Profit_by_State["Profit"])
plt.title("Mean Profit by State")
plt.xlabel("State", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.show()
```

```python
# Horizontal bar chart showing Average Market Spending by State
mean_Marketing_Spend_by_State = df.groupby ("State")["Marketing Spend"].mean()
mean_Marketing_Spend_by_State
print(type(mean_Marketing_Spend_by_State))

mean_Marketing_Spend_by_State = pd.DataFrame(mean_Marketing_Spend_by_State).reset_index()
print(type(mean_Marketing_Spend_by_State))
print(mean_Marketing_Spend_by_State)

plt.barh(mean_Marketing_Spend_by_State["State"],mean_Marketing_Spend_by_State["Marketing Spend"])
plt.title("Average Marketing Spend by State")
plt.xlabel("State", fontsize=12)
plt.ylabel("Marketing Spend", fontsize=12)
plt.show()
```
